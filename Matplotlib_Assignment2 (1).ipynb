# Q2: Create a one line chart using two arrays. The first array represents x and the second array represents y
| x     |     y    |
| ----- | -------- |
| 1    | 1       |
| 2    | 2       |
| 3    | 3       |
| 4    | 5       |
| 5    | 8       |
| 6    | 13      |
| 7    | 20      |

```python
# write your code here ^_^
x = np.array([1,2,3,4,5,6,7])
y= np.array([1,2,3,5,8,13,20])
plt.plot(x, y, c = '0.4')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('X & Y')
```

# Q3: Create a two lines chart using the same previous arrays. 
But this time:
- Mark each point with a circle.
- Use dashed lines.
- Set the size of the marker (circle) to size 10.
- Set the width of the line to 3.
- Add legends. 'Normal' for array x and 'Fast' for array y.

```python
# write your code here ^_^
plt.plot(x, c ='blue',linestyle = 'dashed', linewidth = 3, marker = 'o', markersize = 10, markeredgecolor= 'red')
plt.plot(y, c ='green',linestyle = 'dashed', linewidth = 3, marker = 'o', markersize = 10, markeredgecolor= 'red')
plt.legend(['Normal','Fast'], shadow = True, loc = 0) # location (0 = best)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X & Y')
plt.show()
```

# Q7: Create a pie chart using two arrays. The first array represents python_libraries and the second array represents your skills out of 10 at each library
| python_libraries |   
| ------------ | 
| NumPy        | 
| Pandas       | 
| Matplotlib   | 
| Seaborn      | 
| Plotly       | 
| Scikit-learn | 

- Set a new color for each wedge by using an array of different colors.
- Pull your favorite library wedge 0.3 from the center of the pie.
- Add a shadow to the pie chart

```python
# write your code here ^_^
python_libraries = np.array(['NumPy','Pandas','Matplotlib','Seaborn','Plotly','Sickit-learn'])
SkillRate = np.array([9,8,7,1,8,7])
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(SkillRate,labels=python_libraries, autopct='%1.2f%%', colors =colors,  startangle=180, radius=1.5, shadow=True)
plt.show()
```
